# Testando Regressão Linear Multipla

## Preparando dados

```python
#Copiando dados
dados_preparo01=housing.copy()


#Dropando colunas latitude e longitude
dados_preparo01.drop(columns=['longitude','latitude'], inplace =True)
# dados_preparo01.drop(columns=['total_bedrooms','population','households'],inplace=True)




dados_preparo01.head()
```

### Transformando ocean proximity
4 variáveis diferentes, 1 para cada valor ([NEAR BAY', '<1H OCEAN', 'INLAND', 'NEAR OCEAN', 'ISLAND'])

```python
columns=['NEAR BAY', "<1H OCEAN", 'INLAND', 'NEAR OCEAN', 'ISLAND']

for column in columns:
    conditions=[(dados_preparo01['ocean_proximity']==column), (dados_preparo01['ocean_proximity']!=column)]
    categories=[1,0]
    dados_preparo01[column]=np.select(conditions,categories)

    
dados_preparo01.drop(columns=['ocean_proximity'], inplace=True)
dados_preparo01


```

### Plotando relacao

### Criando Dataset train e test

```python
#Importando biblioteca
from sklearn.model_selection import train_test_split
```

#### Criando Series y e X para as variaveis dependentes e explicativas

```python
y=dados_preparo01['median_house_value']
y
```

```python
X=dados_preparo01.drop(columns=['median_house_value']).copy()
```

#### Criando dataset train e test

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2811)
```

## Regressão Linear com sklearn

#### Setup

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
modelo=LinearRegression()
modelo.fit(X_train,y_train)
```

#### Parametros do Modelos

```python
print("R² = {}".format(modelo.score(X_train,y_train).round(3)))
```

```python
y_predict=modelo.predict(X_test)
print(f'O RMSE do nosso modelo com os dados de teste é {np.sqrt(metrics.mean_squared_error(y_predict, y_test)).round(2)}')


# sklearn.metrics.mean_squared_error
```
